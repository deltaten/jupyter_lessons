### Переименуем колонки

```python
students_performance = students_performance \
        .rename(columns = 
               {'parental level of education': 'parental_level_of_education',
               'test preparation course': 'tes_preparation_course',
                'math score': 'math_score',
                'reading score': 'reading_score',
               'writing score': 'writing_score'})
```

```python
students_performance.columns
```

### Запрос с &(and) и @(программрование переменных)

```python
writing_score_query = 80
```

```python
students_performance.query("writing_score > @writing_score_query & gender == 'female'")
```

```python
score_columns = [i for i in list(students_performance) if 'score' in i] # отбираем имена колонок где присутствует 'score'
```

```python
students_performance.filter(like='score') # тоже самое что выше только через filter
```

```python
sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance) # группирующая переменная gender
```

```python
ax = sns.lmplot(x='math_score', y='reading_score', hue='gender', data=students_performance, fit_reg=False) # без регрессионной прямой
ax.set_xlabels('Math score')
ax.set_ylabels('Reading score')
```
